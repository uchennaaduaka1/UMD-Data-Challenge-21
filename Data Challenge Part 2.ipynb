```python
import requests
import pandas as pd
```

```python
nz_data = requests.get('https://covidmap.umd.edu/api/datesavail?country=New_Zealand&region=Auckland')
df_nz = pd.DataFrame(nz_data.json()['data'])
df_nz
```

```python
de_data = requests.get('https://covidmap.umd.edu/api/datesavail?country=Germany&region=Berlin')
df_de = pd.DataFrame(de_data.json()['data'])
df_de
```

```python
tk_data = requests.get('https://covidmap.umd.edu/api/datesavail?country=Australia&region=Queen%')
df_tk = pd.DataFrame(tk_data.json()['data'])
df_tk
```

```python
bz_data = requests.get('https://covidmap.umd.edu/api/datesavail?country=Brazil&region=Rio_de_Jan%')
df_bz = pd.DataFrame(bz_data.json()['data'])
df_bz
```

```python
url1 = 'https://covidmap.umd.edu/api/resources?indicator=mask&type=daily&country=New_Zealand&daterange=20200501-20210201'
nz_masks = requests.get(url1)
df_nz_masks = pd.DataFrame(nz_masks.json()['data'])
df_nz_masks
```

```python
url2 = 'https://covidmap.umd.edu/api/resources?indicator=mask&type=daily&country=Germany&region=Berlin&daterange=20200501-20210201'
de_masks = requests.get(url2)
df_de_masks = pd.DataFrame(de_masks.json()['data'])
df_de_masks
```

```python
url3 = 'https://covidmap.umd.edu/api/resources?indicator=mask&type=daily&country=Canada%&region=Ontario%&daterange=20200501-20210201'
ca_masks = requests.get(url3)
df_ca_masks = pd.DataFrame(ca_masks.json()['data'])
df_ca_masks
```

```python
url4 = 'https://covidmap.umd.edu/api/resources?indicator=mask&type=daily&country=Brazil&region=Rio_de_J%&daterange=20200501-20210201'
bz_masks = requests.get(url4)
df_bz_masks = pd.DataFrame(bz_masks.json()['data'])
df_bz_masks
```

```python
covid_url1 = 'https://covidmap.umd.edu/api/resources?indicator=covid&type=daily&country=Brazil&region=Rio_de_J%&daterange=20200501-20210201'
bz_covid = requests.get(covid_url1)
df_bz_covid = pd.DataFrame(bz_covid.json()['data'])
df_bz_covid
```

```python
covid_url2 = 'https://covidmap.umd.edu/api/resources?indicator=covid&type=daily&country=Canada&region=Ontario%&daterange=20200501-20210201'
ca_covid = requests.get(covid_url2)
df_ca_covid = pd.DataFrame(ca_covid.json()['data'])
df_ca_covid
```

```python
covid_url3 = 'https://covidmap.umd.edu/api/resources?indicator=covid&type=daily&country=Germany&region=Berlin%&daterange=20200501-20210201'
de_covid = requests.get(covid_url3)
df_de_covid = pd.DataFrame(de_covid.json()['data'])
df_de_covid
```

```python
covid_url4 = 'https://covidmap.umd.edu/api/resources?indicator=covid&type=daily&country=Japan&region=Tokyo%&daterange=20200501-20210201'
jp_covid = requests.get(covid_url4)
df_jp_covid = pd.DataFrame(jp_covid.json()['data'])
df_jp_covid
```

```python
url5 = 'https://covidmap.umd.edu/api/resources?indicator=mask&type=daily&country=Japan%&region=Tokyo%&daterange=20200501-20210201'
jp_masks = requests.get(url5)
df_jp_masks = pd.DataFrame(jp_masks.json()['data'])
df_jp_masks
```

## Actual datasets

```python
df_ge = df_de_masks.merge(df_de_covid, left_on='survey_date', right_on='survey_date')
df_ge_final = df_ge[['country_x', 'region_x','percent_mc', 'percent_cli', 'survey_date']]
df_ge_final.to_csv('Germany_final_data.csv')
```

```python
df_ca = df_ca_masks.merge(df_ca_covid, left_on='survey_date', right_on='survey_date')
df_ca_final = df_ca[['country_x', 'region_x','percent_mc', 'percent_cli', 'survey_date']]
df_ca_final.to_csv('Canada_final_data.csv')
```

```python
df_bz = df_bz_masks.merge(df_bz_covid, left_on='survey_date', right_on='survey_date')
df_bz_final = df_bz[['country_x', 'region_x','percent_mc', 'percent_cli', 'survey_date']]
df_bz_final.to_csv('Brazil_final_data.csv')
```

```python
df_jp = df_jp_masks.merge(df_jp_covid, left_on='survey_date', right_on='survey_date')
df_jp_final = df_jp[['country_x', 'region_x','percent_mc', 'percent_cli', 'survey_date']]
df_jp_final.to_csv('Japan_final_data.csv')
```
