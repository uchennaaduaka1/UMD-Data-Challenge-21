```python
import requests
import pandas as pd
```

Find all the countries in the COVID Map API

```python
all_count = requests.get("https://covidmap.umd.edu/api/country").json()
```

```python
df_count = pd.DataFrame(all_count['data'])
df_count.head()
```

```python
#Make the dataframe into a csv to view in Excel
df_count.to_csv('countries.csv')
```

Find all the regions in the COVID MAP API

```python
all_reg = requests.get("https://covidmap.umd.edu/api/region").json()
df_reg = pd.DataFrame(all_reg['data'])
```

```python
df_reg.head()
```

```python
df_reg.to_csv('regions_covid.csv')
```

It appears that there are a number of specific regions but some countries that we have listed above don't have regions associated with them (ex. United States). For our analysis, we want to look at what time periods are avaialable. 

```python
all_dates = requests.get('https://covidmap.umd.edu/api/datesavail?country=%').json()
df_alldates = pd.DataFrame(all_dates['data'])
df_alldates.head()
```

```python
#Find how many dates there are for each country using the groupby function
count_dates = df_alldates.groupby(['country']).count()
```

```python
count_dates
```

```python
count_dates.describe()
#Describe tells us that the average country has 264 dates of data collection
#However it seems alot of countries have 308 dates
```

```python
dates_308 = count_dates[count_dates.loc[:,'survey_date']==308]
```

```python
#List of dates that have 308 survey collection dates
dates_308.head()
```

```python
print(f"There are {len(dates_308)} countries with 308 survey collection dates")
```

```python
dates_308.to_csv('dates_308.csv')
```
