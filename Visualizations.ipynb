```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
pd.set_option("display.max.columns", None)
Germany_df = pd.read_csv('Germany_final_data.csv')
Canada_df = pd.read_csv('Canada_final_data.csv')
Japan_df = pd.read_csv('Japan_final_data.csv')
Brazil_df = pd.read_csv('Brazil_final_data.csv')
```

```python
Germany_ave_percent_mc = Germany_df["percent_mc"].mean()
Japan_ave_percent_mc = Japan_df["percent_mc"].mean()
Canada_ave_percent_mc = Canada_df["percent_mc"].mean()
Brazil_ave_percent_mc = Brazil_df["percent_mc"].mean()
Germany_ave_percent_cli = Germany_df["percent_cli"].mean()
Japan_ave_percent_cli = Japan_df["percent_cli"].mean()
Canada_ave_percent_cli = Canada_df["percent_cli"].mean()
Brazil_ave_percent_cli = Brazil_df["percent_cli"].mean()
regions = ["Berlin","Tokyo","Ontario","Rio de Jainero"]
ave_percent_mc = [Germany_ave_percent_mc,Japan_ave_percent_mc,Canada_ave_percent_mc,Brazil_ave_percent_mc]
ave_percent_cli = [Germany_ave_percent_cli,Japan_ave_percent_cli,Canada_ave_percent_cli,Brazil_ave_percent_cli]
```

```python
average_mc = plt.bar(regions,ave_percent_mc)
average_mc = plt.xlabel('Regions')
average_mc = plt.ylabel('Percent of Mask Wearing')
```

```python
average_cli = plt.bar(regions,ave_percent_cli)
average_mc = plt.xlabel('Regions')
average_mc = plt.ylabel('Percent of No Mask Wearing')
```
